```python
class StreamListener(tweepy.StreamListener):
    def on_status(self, status):
        try:
            if status.retweeted_status:
                return
        except:
            name = status.user.screen_name
            text = status.text
            created = status.created_at
            print(text)
            w.writerow({'time': created, 'user': name, 'text': text})
    def on_error(self, status_code):
        if status_code == 420:
            return False
```

```python
stream_listener = StreamListener()
# Subjectivity: How subjective, or opinionated a word is. 0 is fact. +1 is very much opinion.
stream = tweepy.Stream(auth=api.auth, listener=stream_listener)
stream.filter(track=["@realdonaldtrump"], is_async=True)
time.sleep(collection_time)
stream.disconnect()
target.close()
```
